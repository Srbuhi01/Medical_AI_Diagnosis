```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

resnet_log = '../models/training_log_cnn.csv'
swin_log = '../models/swin1_training_log.csv'

df_resnet = pd.read_csv(resnet_log)
df_swin = pd.read_csv(swin_log)

plt.figure(figsize=(12, 7))
sns.set_style("whitegrid")

# ResNet (Blue)
plt.plot(df_resnet['epoch'], df_resnet['val_loss'],
         label='ResNet50 (CNN)', marker='o', linestyle='-', color='blue', linewidth=2)

# Swin (Red)
plt.plot(df_swin['epoch'], df_swin['val_loss'],
         label='Swin Transformer (ViT)', marker='s', linestyle='--', color='red', linewidth=2)

plt.title('Model Comparison: Validation Loss', fontsize=16)
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Loss (Lower is Better)', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)

# Ցույց ենք տալիս
plt.show()
```
